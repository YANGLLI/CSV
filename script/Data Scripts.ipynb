```python
# Electives Data Script
import pandas as pd
import numpy as np
from collections import defaultdict
import random

align_course = {'CS 5001': 'Intensive Foundations of CS',
                'CS 5002': 'Discrete and Data Structures',
                'CS 5004': 'Object-Oriented Design', 
                'CS 5006': 'Algorithms', 
                'CS 5007': 'Computer Systems'}

student_df = pd.read_csv('Students.csv')
course_df = pd.read_csv('Courses.csv')
course_id = course_df['CourseId'].values
course_name = course_df['CourseName'].values
all_cs_course = dict(zip(course_id, course_name))

def prepare_course(year):  # ignore the effect of term on the number of courses taken
    ret = align_course.copy()
    if year == 2017:
        selected_electives = {}
    elif year == 2016:
        selected_electives = random.sample(all_cs_course.items(), 4)
    else:
        selected_electives = random.sample(all_cs_course.items(), 7)
    
    ret.update({course[0]: course[1] for course in selected_electives})
    return ret

csv_data = defaultdict(list)
for i in range(400):
    year = student_df['EntryYear'][i]
    course_taken = prepare_course(year)
    course_cnt = len(course_taken)
    csv_data['NeuId'].extend(['%03d' % i] * course_cnt)
    csv_data['CourseId'].extend(course_taken.keys())
    csv_data['CourseName'].extend(course_taken.values())
    csv_data['Term'].extend([random.choice(['SPRING', 'FALL']) for _ in range(course_cnt)])
    csv_data['Year'].extend([year] * course_cnt)
    csv_data['GPA'].extend([random.choice(['A', 'A-', 'B+', 'B', 'B-', 'C+']) for _ in range(course_cnt)])
    csv_data['Retake'].extend(['No'] * course_cnt)
    csv_data['Plagiarism'].extend(['No'] * course_cnt)

pd.DataFrame(csv_data).to_csv('Electives.csv', index=False)
```
